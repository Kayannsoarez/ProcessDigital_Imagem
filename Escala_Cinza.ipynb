```python
# Importando as bibliotecas
import cv2
from matplotlib import pyplot as plt
```

 Criando função para facilitar a exibição das imagens:

```python
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## **Carregando Imagem**

```python
# Carregando a imagem
image = cv2.imread('/content/castara.jpeg')

imshow("Castara, Tobago", image)
```

## **Escala de Cinza**

```python
# Convertendo em Tons de Cinza
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow("Convertido em Escala de Cinza", gray_image)
```

## **Dimensões da imagem em escala de cinza**

```python
gray = cv2.imread('/content/gray.png')

imshow("Gray - Map", gray)
```
