```python
# Importando as Bibiotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Função de Exibição
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## **Operações Aritmeticas**

```python
# Carregando Imagem em Niveis de Cinza
image = cv2.imread('/content/liberty.jpeg', 0)
imshow("Imagem em tons de cinza",image)

# Criando Matriz de Operação
M = np.ones(image.shape, dtype = "uint8") * 100 
```

```python
# Analizando dimensões

print("Dimensão Imagem\n", image, "\n")
print("Dimensão Matriz\n", M)
```

## **Aumentando o brilho**

```python
# Usamos a matriz M, para aumentar o brlho da imagem
added = cv2.add(image, M)
imshow("Aumentando o Brilho", added)

# Somando o Aumento
added2 = image + M 
imshow("Adição", added2)
```

## **Diminuindo o brilho**

```python
# Diminuindo o brilho
subtracted = cv2.subtract(image, M)
imshow("Redução do Brilho", subtracted)

# Subtraindo a redução
subtracted = image - M 
imshow("Subtração do Brilho", subtracted)
```

## **Operação de Bitwise e Masking**

```python
# Criando um Quadrado
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
imshow("Quadrado", square)

# Criando uma Elipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (160, 120), (150, 150), 30, 0, 180, 255, -1)
imshow("Elipse", ellipse)
```

```python
# Conjunção
And = cv2.bitwise_and(square, ellipse)
imshow("Operação AND", And)

# Disjunção
bitwiseOr = cv2.bitwise_or(square, ellipse)
imshow("Operação OR", bitwiseOr)

# Disjunção Exclusiva
bitwiseXor = cv2.bitwise_xor(square, ellipse)
imshow("Operção XOR", bitwiseXor)

# Negação
bitwiseNot_sq = cv2.bitwise_not(square)
imshow("Operação NOT", bitwiseNot_sq)
```
