```python
# Importando Biblioteca do OpenCV
import cv2
```

```python
# Conferindo a versão de execução
print(cv2.__version__)
```

## **Carregando Imagem**


```python
# Carregando a imagem usando 'imread'
praia = cv2.imread('/content/castara.jpeg')
flores = cv2.imread('/content/flowers.jpeg')
```

## **Exibindo Imagens**

```python
from matplotlib import pyplot as plt

# Exibindo a imagem com matplotlib
plt.imshow(cv2.cvtColor(praia, cv2.COLOR_BGR2RGB))
plt.show()
```

Criando uma função para facilitar a exibição das imagens: 


```python
def imshow(title = "", image = None):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Testando
imshow("Exibindo Imagem", flores)
```

## **Salvando Imagem**


```python
# Para salvar especifique o nome do arquivo e a imagem a ser salva
cv2.imwrite('output.jpg', flores)
```

```python
# Salvando como PNG (Portable Network Graphics)
cv2.imwrite('output.png', flores)
```
