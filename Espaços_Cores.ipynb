```python
# Importando as Bibliotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Função de Exibição
def imshow(title = "Image", image = None, size = 10):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Carregando imagem
image = cv2.imread('/content/view.jpeg')
```

```python
# Obtendo os espaços de cores separadamente
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
```

```python
imshow("Somente Canal Azul", B)
```

O espaço de cor resultante parecerá uma escala de cinza, pois não possui os outros canais de cores

## **Formato RGB**

```python
zeros = np.zeros(image.shape[:2], dtype = "uint8")

imshow("Red", cv2.merge([zeros, zeros, R]))
imshow("Green", cv2.merge([zeros, G, zeros]))
imshow("Blue", cv2.merge([B, zeros, zeros]))
```

```python
B, G, R = cv2.split(image)

# Juntando os Canais 
merged = cv2.merge([B, G, R]) 
imshow("RGB Unificado", merged) 
```

```python
# Amplificando o Azul
merged = cv2.merge([B+100, G, R])
imshow("Blue Boost", merged)
```

## **Formato HSV**

```python
# Carregando imagem
hsv = cv2.imread('/content/HSV_color.jpg')
imshow("HSV", hsv)
```

- Hue: 0 - 179 
- Saturation: 0 - 255
- Value (Intensity): 0 - 255

```python
# Convertendo para HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
imshow('HSV', hsv_image)
```

```python
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_HSV2RGB))
plt.show()
```

```python
imshow("Hue", hsv_image[:, :, 0])
imshow("Saturation", hsv_image[:, :, 1])
imshow("Value", hsv_image[:, :, 2])
```
