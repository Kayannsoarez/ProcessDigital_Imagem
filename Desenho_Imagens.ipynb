```python
# Importando as Bibiotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Função de Exibição
def imshow(title = "Image", image = None, size = 8):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## **Desenho**

```python
# Criando Imagem Preta - RGB
image = np.zeros((512,512,3), np.uint8)

# Criando Imagem Preta - Escala de Cinza
image_gray = np.zeros((512,512), np.uint8)

imshow("RGB Color", image)
imshow("Grayscale", image_gray)
```

## **Desenho - Linha**

```python
cv2.line(image, (0,0), (511,511), (255,127,0), 5)

imshow("Linha Diagonal", image)
```

## **Desenho - Retangulo**

```python
# Criando Imagem Preta
image = np.zeros((512,512,3), np.uint8)

# Criando Linhas do Quadrado
cv2.rectangle(image, (100,100), (300,250), (127,50,127), 10)
imshow("Retangulo", image)
```

## **Desenho - Circulo**

```python
# Criando Imagem Preta
image = np.zeros((512,512,3), np.uint8)

# cv2.cirlce(image, center, radius, color, fill)
cv2.circle(image, (256, 256), 120, (15,150,50), -1) 
imshow("Circulo", image)
```

## **Desenho - Polígono**

```python
# Criando Imagem Preta
image = np.zeros((512,512,3), np.uint8)

# Definindo Pontos
pts = np.array( [[10,50], [400,50], [90,200], [50,500]], np.int32)

# remodelar os pontos exigido pelas polilinhas
pts = pts.reshape((-1,1,2))

# cv2.polylines(image, points, Closed?, color, thickness)
cv2.polylines(image, [pts], True, (0,0,255), 3)
imshow("Polígono Vermelho", image)
```

## **Desenho - Texto**

```python
# Criando Imagem Preta
image = np.zeros((512,512,3), np.uint8)

# Texto
Stryng =  'Hello World!'

# cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)
cv2.putText(image, Stryng, (50,256), cv2.FONT_HERSHEY_SCRIPT_COMPLEX, 2.5, (40,200,0), 4)
imshow("Texto Desenhado", image)
```
