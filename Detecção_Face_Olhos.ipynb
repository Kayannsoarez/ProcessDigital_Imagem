```python
# Importando as Bibiotecas
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Função de Exibição
def imshow(title = "Image", image = None, size = 8):
    w, h = image.shape[0], image.shape[1]
    aspect_ratio = w/h
    plt.figure(figsize=(size * aspect_ratio,size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

## **Detecção Facial usando Classificador HaarCascade**

Este metodo usa uma série de classificadores (cascata) para identificar objetos em uma imagem. Eles são treinados para identificar um tipo de objeto, porém, podemos usar vários deles em paralelo. Os classificadores HAAR são treinados usando muitas imagens positivas (imagens com o objeto presente) e imagens negativas (imagens sem o objeto presente).

```python
# Carrengando os Modelos de Classificadores
face_classifier = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
eye_classifier = cv2.CascadeClassifier('/content/haarcascade_eye.xml')
```

```python
# Carregando imagem e Convertendo em Escala de Cinza
image = cv2.imread('/content/ElliotAlderson.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Deteção de Faces
faces = face_classifier.detectMultiScale(gray, scaleFactor = 1.3, minNeighbors = 5)

# Caso para nenhuma face detectada
if faces is ():
    print("Nenhuma Face Encontrada")

# Busca por faces na imagem e desenho do retangulo de marcação
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (127,0,255), 2)

imshow('Face Detectada', image)
```

```python
# Carregando imagem e Convertendo em Escala de Cinza
image = cv2.imread('/content/Shrek.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Deteção de Faces
faces = face_classifier.detectMultiScale(gray, scaleFactor = 1.3, minNeighbors = 5)

# Caso para nenhuma face detectada
if faces is ():
    print("Nenhuma Face Encontrada")

# Busca por faces na imagem e desenho do retangulo de marcação
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (127,0,255), 2)

imshow('Face Detectada', image)
```

## **Detecção Olhos e Faces - Classificador HaarCascade**

```python
# Carregando imagem e Convertendo em Escala de Cinza
img = cv2.imread('/content/ElliotAlderson.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Deteção de Faces
faces = face_classifier.detectMultiScale(gray, scaleFactor = 1.3, minNeighbors = 5)

# Caso para nenhuma face detectada
if faces is ():
    print("Nenhuma Face Encontrada")

for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(127,0,255),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(roi_gray, 1.2, 3)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(255,255,0),2)

imshow('Olhos e Faces Detecadas', img)
```
